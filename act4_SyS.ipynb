```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
# Parámetros básicos
fs = 10000              # Frecuencia de muestreo (Hz)
t = np.linspace(0, 1, fs)  # Tiempo de 0 a 1 segundo con fs muestras

# Señal del mensaje (baja frecuencia): onda senoidal de 50 Hz
fm = 50                 # Frecuencia del mensaje (Hz)
Am = 1                  # Amplitud del mensaje
mensaje = Am * np.sin(2 * np.pi * fm * t)

# Graficar la señal de entrada (mensaje)
plt.figure(figsize=(10, 4))
plt.plot(t, mensaje)
plt.title("Señal del Mensaje (Información a transmitir)")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Señal portadora (alta frecuencia)
fc = 1000               # Frecuencia de la portadora (Hz)
Ac = 1                  # Amplitud de la portadora
portadora = Ac * np.cos(2 * np.pi * fc * t)

# Modulación AM: (1 + señal de mensaje) * portadora
senal_am = (1 + mensaje) * portadora

# Graficar señal AM en el tiempo
plt.figure(figsize=(10, 4))
plt.plot(t, senal_am)
plt.title("Señal Modulada en Amplitud (AM)")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# FFT para obtener el espectro de frecuencias
N = len(senal_am)
f = fftfreq(N, 1/fs)
senal_am_fft = np.abs(fft(senal_am)) / N

# Graficar espectro
plt.figure(figsize=(10, 4))
plt.plot(f[:N//2], senal_am_fft[:N//2])  # Solo parte positiva del espectro
plt.title("Espectro de Frecuencia de la Señal AM")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Magnitud")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Agregar ruido blanco (media = 0, desviación estándar = 0.5)
ruido = np.random.normal(0, 0.5, size=senal_am.shape)
senal_ruidosa = senal_am + ruido

# Graficar señal con ruido
plt.figure(figsize=(10, 4))
plt.plot(t, senal_ruidosa)
plt.title("Señal AM con Ruido Blanco")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Simular distorsión no lineal (efecto cuadrático)
senal_distorsionada = senal_am + 0.3 * senal_am**2

# Graficar señal distorsionada
plt.figure(figsize=(10, 4))
plt.plot(t, senal_distorsionada)
plt.title("Señal AM con Distorsión No Lineal")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Atenuar la señal al 50% de su amplitud
senal_atenuada = 0.5 * senal_am

# Graficar señal atenuada
plt.figure(figsize=(10, 4))
plt.plot(t, senal_atenuada)
plt.title("Señal AM Atenuada (50%)")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid(True)
plt.tight_layout()
plt.show()
```
